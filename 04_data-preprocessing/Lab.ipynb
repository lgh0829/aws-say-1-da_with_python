5)	토요일 환승 인원이 100000명 이상의 역의 역명과 다른 요일 환승인원 출력

```python
import pandas as pd
subway = pd.read_csv('data/subway.csv')
```

```python
# 1) 5행만 출력
subway.head()
```

```python
# 2) 행 제목만 출력
subway.index
```

```python
# 3) 열 제목만 출력
subway.columns
```

```python
# 4) “평일” 환승인원을 기준으로 오름차순 정렬하여 출력
subway.sort_values(by='평일', ascending=True)
```

```python
# 5) 토요일 환승 인원이 100000명 이상의 역의 역명과 다른 요일 환승인원 출력
result = subway[subway['토요일'] >= 100000]
result.loc[:,('역명', '평일', '일요일')]
```

```python
# 6) 일요일 환승 인원만 역명과 함께 출력
subway.loc[:, ('역명', '일요일')]
```

```python
# 7) 11~20번째 행의 일요일 환승 인원만 역명과 함께 출력
subway.loc[11:21, ('역명', '일요일')]
```

```python
# 8) 역명만 출력. 단 중복해서 출력하지는 말 것
# 데이터셋 자체에 중복되는 역명이 있지 않으므로 역명만 출력함
subway.loc[:,'역명']

subway['역명'].unique() # 의도
```
