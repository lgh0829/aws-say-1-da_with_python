```python
# wikidocs의 html 파일 수집
import urllib.request

d = urllib.request.urlopen("https://comic.naver.com/webtoon/list?titleId=747269")

data = d.read().decode("utf-8")

print(data)
```

```python
# 한글 단어만 추출
import re  
 
all_hangule = re.findall("[가-힣]+", data) 

for i in all_hangule   : 
    print (i) 
```

```python
# http 헤더 추출
status =  d.getheaders()  

for s in status   : 
    print(s) 
```

```python
# 검색할 키워 준비. 한글 키워드는 url에 받을 수 없기 때문에 utf-8 인코딩이 필요함
# urllib.parse의 quote 메서드는 한글 키워드를 입력하면 utf-8 인코딩 변환을 제공함
import urllib.parse 

a = urllib.parse.quote("대선") 
print(a)
```

```python
# 입력받은 키워드를 네이버에 검색한 결과 가져오기

import urllib.parse 
import urllib.request
import re

keyword = input("검색할 단어를 입력하세요")
keyword = urllib.parse.quote("utf-8")

url = "https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query="+keyword 
d = urllib.request.urlopen(url)

print ( d.status )

if d.status == 200  :
    data = d.read().decode("utf-8 ") 
    
    f = open ("result.html","w")
    f.write (data)
    f.close()

    # html에 한글 데이터만 파싱
    all_hangule = re.findall("[가-힣]+", data)

    for i in all_hangule: 
        print(i)
```
