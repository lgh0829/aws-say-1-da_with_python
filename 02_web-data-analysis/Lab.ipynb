```python
### Lab 1-1

import urllib.request
import urllib.parse
from bs4 import BeautifulSoup

# html request
url = urllib.request.urlopen("https://sports.news.naver.com/index")

# after health check, pull html file
if url.status == 200:
    data = url.read().decode("utf-8")

    file = open("test.html","w")
    file.write(data)
    file.close()

# ready to html parsing
data = BeautifulSoup(data,'html.parser')

# parsing headlines and links
headlines = [headline['title'] for headline in data.find_all("a", title=True)]
links = [headline['href'] for headline in data.find_all("a", title=True)]

for idx in range(len(headlines)):
    print(f"[{idx}] {headlines[idx]}\n{links[idx]}\n")
```

```python
### Lab 1-2
import urllib.request
import urllib.parse
from bs4 import BeautifulSoup

# html request
keyword = urllib.parse.quote("폭싹 속았수다 리뷰")
url = urllib.request.urlopen("https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query=" + keyword)

# after health chekck, pull html file
if url.status == 200:
    data = url.read().decode("utf-8")
    
    file = open("test2.html","w")
    file.write(data)
    file.close()

# ready to html parsing
data = BeautifulSoup(data, 'html.parser')

# parsing titles, descriptions and links
titles = [a.get_text(strip=False) for a in data.find_all('a', class_='title_link')]
links = [a['href'] for a in data.find_all('a', class_='title_link')]
descrips = [a.get_text(strip=False) for a in data.find_all('a', class_='dsc_link')]

for idx in range(len(titles)):
    print(f"[{idx}] {titles[idx]}\n{descrips[idx]}\n{links[idx]}\n")
```
