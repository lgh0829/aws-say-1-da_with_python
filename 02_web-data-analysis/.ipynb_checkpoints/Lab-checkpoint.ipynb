```python
### Lab 1-1

import urllib.request
import urllib.parse
from bs4 import BeautifulSoup

# html request
url = urllib.request.urlopen("https://sports.news.naver.com/index")

# after health check, pull html file
if url.status == 200:
    data = url.read().decode("utf-8")

    file = open("test.html","w")
    file.write(data)
    file.close()

# ready to html parsing
data = BeautifulSoup(data,'html.parser')

# parsing headlines and links
headlines = [headline['title'] for headline in data.find_all("a", title=True)]
links = [headline['href'] for headline in data.find_all("a", title=True)]

for idx in range(len(headlines)):
    print(f"[{idx}] {headlines[idx]}\n{links[idx]}\n")
```

```python
### Lab 1-2
import urllib.request
import urllib.parse
from bs4 import BeautifulSoup

# html request
keyword = urllib.parse.quote("폭싹 속았수다 리뷰")
url = urllib.request.urlopen("https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query=" + keyword)

# after health chekck, pull html file
if url.status == 200:
    data = url.read().decode("utf-8")
    
    file = open("test2.html","w")
    file.write(data)
    file.close()

# ready to html parsing
data = BeautifulSoup(data, 'html.parser')

# parsing titles, descriptions and links
titles = [a.get_text(strip=False) for a in data.find_all('a', class_='title_link')]
links = [a['href'] for a in data.find_all('a', class_='title_link')]
descrips = [a.get_text(strip=False) for a in data.find_all('a', class_='dsc_link')]

for idx in range(len(titles)):
    print(f"[{idx}] {titles[idx]}\n{descrips[idx]}\n{links[idx]}\n")
```

# Lab2

naver 부동산(https://land.naver.com/)  의 오늘의 뉴스 중 첫번째 뉴스를  클릭하여 뉴스의 상세 내용을
크롤링하여 워드 클라우드를 만들어보세요. ( Selenium 과 BeautifulSoup 이용 반드시 Selenium 을 써야하는 것은 아니지만 .. )

```python
import urllib.request
import urllib.parse
from bs4 import BeautifulSoup
from selenium import webdriver
import chromedriver_autoinstaller
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

chromedriver_autoinstaller.install()  # Check if the current version of chromedriver exists
                                      # and if it doesn't exist, download it automatically,
                                      # then add chromedriver to path

# search today news
driver = webdriver.Chrome()
driver.get("https://land.naver.com/")
original_tab = driver.current_window_handle

news_entry = driver.find_element(By.XPATH, "/html/body/div[4]/div[3]/div[3]/div[1]/div[1]/div/h3/a")
news_entry.click()

# change tab
tabs = driver.window_handles

for tab in tabs:
    if tab != original_tab:
        driver.switch_to.window(tab)
        break

# change url
change_url = driver.current_url
```

```python
# parsing
req = urllib.request.urlopen(change_url)

if req.status == 200:
    data = req.read().decode("utf-8")

data = BeautifulSoup(data, 'html.parser')

scripts = data.find('div', id='newsct')


headline = scripts.find('h2').get_text(strip=False)
article = [item.get_text(strip=False) for item in scripts.find_all('article', id='dic_area')]
text = headline[0] + article[0]

```

```python
# 자연어 한국어 분석
from konlpy.tag import Okt 

okt = Okt()

tokens = okt.nouns(text)
```

```python
# Word Cloud
import matplotlib.pyplot as plt
%matplotlib inline

from wordcloud import WordCloud, STOPWORDS
from collections import Counter

wordcloud = WordCloud(max_font_size=200,
                     font_path="/System/Library/fonts/AppleSDGothicNeo.ttc",  # Mac OS
                     background_color='#ffffff',
                      max_words = 2000,
                     colormap="rainbow",
                     width=1200, height=800).generate_from_text(text)

plt.figure(figsize=(10,8))
plt.imshow(wordcloud)
plt.tight_layout(pad=0)
plt.axis('off')

plt.savefig('wordcloud_news.png', bbox_inches='tight')
plt.show()
```

# Lab 3

무신사에서 운동화를 검색해서 나온 페이지에서 운동화 정보를 출력 (가격, 상품명 등 )

```python
from selenium import webdriver
import chromedriver_autoinstaller

chromedriver_autoinstaller.install()  # Check if the current version of chromedriver exists
                                      # and if it doesn't exist, download it automatically,
                                      # then add chromedriver to path

driver = webdriver.Chrome()
driver.get("https://www.musinsa.com/")


from selenium.webdriver.common.by import By
search_entry = driver.find_element(By.XPATH, """//*[@id="commonLayoutHeader"]/section/div[2]/div/button[1]""")

search_entry.click()

keyword_element = driver.find_element(By.XPATH, """//*[@id="commonSearchHome"]/div/main/header/div/input""")
keyword_element.clear()
keyword_element.send_keys("운동화")

from selenium.webdriver.common.keys import Keys
keyword_element.send_keys(Keys.ENTER)

# current url
current_url = driver.current_url
```

```python
# keyword에 대한 정보 출력

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

try:
    # waiting
    elements = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located(
            (By.CLASS_NAME, "sc-fpgwy.lgruTi")
        )
    )
    
    items = [item.text for item in elements]
    
    brand = []
    prod_name = []
    price =[]
    
    for item in items:
        lines = item.strip().split('\n')
        
        brand.append(lines[0])
        prod_name.append(lines[1])
        price.append(lines[2])
    
finally:
    
    for idx in range(len(brand)):
        print(f"[{idx}] {brand[idx]}\n{prod_name[idx]}\n{price[idx]}\n")
```
